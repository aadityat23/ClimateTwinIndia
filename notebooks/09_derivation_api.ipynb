```python
MODEL = "gemini-2.0-flash"
```

```python
import time

def ask(prompt):

    for attempt in range(5):

        try:

            response = client.models.generate_content(
                model=MODEL,
                contents=prompt
            )

            return response.text

        except Exception as e:

            print(f"Retry {attempt+1}/5")

            print(e)

            time.sleep(10)

    return None
```

```python
text = ask(df.loc[0,"Prompt"])

print(text)
```
