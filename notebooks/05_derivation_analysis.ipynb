```python
import numpy as np

master["normalized_margin"] = np.nan

# ---------------- Persistent ----------------

mask = master["subcategory"]=="persistent_regional_anomaly_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"min_yearly_mean_abs_anomaly"]
        -
        master.loc[mask,"support_mean_abs_p75"]
    ).abs()
) / (
    master.loc[mask,"support_mean_abs_p75"]
    -
    master.loc[mask,"refute_mean_abs_p50"]
).abs()


# ---------------- Localized ----------------

mask = master["subcategory"]=="localized_intensification_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"localized_contrast_change"]
        -
        master.loc[mask,"support_contrast_change_q75"]
    ).abs()
) / (
    master.loc[mask,"support_contrast_change_q75"]
    -
    master.loc[mask,"refute_contrast_change_q25"]
).abs()


# ---------------- Wet ----------------

mask = master["subcategory"]=="wet_anomaly_consistency_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"fraction_cells_rainfall_increased"]
        -
        master.loc[mask,"support_fraction_increase_q75"]
    ).abs()
) / (
    master.loc[mask,"support_fraction_increase_q75"]
    -
    master.loc[mask,"refute_fraction_increase_q25"]
).abs()


# ---------------- Compound ----------------

mask = master["subcategory"]=="compound_state_transition_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"regional_temperature_change"]
        -
        master.loc[mask,"support_temperature_change_q66"]
    ).abs()
) / (
    master.loc[mask,"support_temperature_change_q66"]
    -
    master.loc[mask,"refute_temperature_change_q33"]
).abs()


# ---------------- Spatial ----------------

mask = master["subcategory"]=="spatial_coherence_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"largest_component_size"]
        -
        master.loc[mask,"support_largest_component_min"]
    ).abs()
) / (
    master.loc[mask,"support_largest_component_min"]
    -
    master.loc[mask,"refute_largest_component_max"]
).abs()


# ---------------- Reliability ----------------

mask = master["subcategory"]=="reliability_aware_claim_verification"

master.loc[mask,"normalized_margin"] = (
    (
        master.loc[mask,"mean_abs_anomaly"]
        -
        master.loc[mask,"support_mean_abs_q85"]
    ).abs()
) / (
    master.loc[mask,"support_mean_abs_q85"]
    -
    master.loc[mask,"refute_mean_abs_q50"]
).abs()

display(master[["id","normalized_margin"]].head())
```

```python
import re
import pandas as pd

def parse_reasoning(text):

    text=str(text)

    metric1=None
    metric2=None
    metric3=None
    decision=None

    patterns=[
        r"Metric[_ ]?1\s*[:=]\s*([\-0-9.]+)",
        r"Metric[_ ]?2\s*[:=]\s*([\-0-9.]+)",
        r"Metric[_ ]?3\s*[:=]\s*([\-0-9.]+)"
    ]

    for i,p in enumerate(patterns):

        m=re.search(p,text,re.I)

        if m:

            value=float(m.group(1))

            if i==0:
                metric1=value

            elif i==1:
                metric2=value

            else:
                metric3=value

    m=re.search(r"Decision\s*[:=]\s*([ABC])",text,re.I)

    if m:
        decision=m.group(1).upper()

    return metric1,metric2,metric3,decision
```

```python
import re

with open("../experiments/experiment_05_derivation\\raw_outputs\\deepseek_derivation.txt","r",encoding="utf8") as f:
    txt=f.read()

pattern=r"Metric_1 = ([\-0-9\.]+).*?Metric_2 = ([\-0-9\.]+).*?Metric_3 = ([\-0-9\.]+).*?Decision = ([ABC])"

matches=re.findall(pattern,txt,re.S)

print("Found:",len(matches))

pred=[]

for i,m in enumerate(matches[:40]):

    pred.append({

        "id":f"HYP-{i+1:03d}",
        "Pred_Metric1":float(m[0]),
        "Pred_Metric2":float(m[1]),
        "Pred_Metric3":float(m[2]),
        "Pred_Decision":m[3]

    })

pred=pd.DataFrame(pred)

display(pred.head())
```

```python
def recover_persistent(row):

    # SUPPORT
    if (
        row["Pred_Metric1"] >= row["support_mean_abs_p75"]
        and
        row["Pred_Metric2"] >= row["support_fraction_above_p75"]
        and
        row["Pred_Metric3"] >= row["support_completeness_p10"]
    ):
        return "C"

    # REFUTE
    elif (
        row["Pred_Metric1"] <= row["refute_mean_abs_p50"]
        or
        row["Pred_Metric2"] <= row["refute_fraction_above_p75"]
        or
        row["Pred_Metric3"] < row["support_completeness_p10"]
    ):
        return "A"

    # OTHERWISE

    return "B"


df["RecoveredDecision"] = df.apply(
    recover_persistent,
    axis=1
)
```

```python
df["RecoveredCorrect"] = (
    df["RecoveredDecision"] ==
    df["GT_Decision"]
)

print(
    "Recovered Accuracy:",
    df["RecoveredCorrect"].mean()*100
)
```

```python
def category_accuracy(df):

    return (
        df.groupby("subcategory")
          .apply(
              lambda x: (
                  x["RecoveredDecision"] ==
                  x["GT_Decision"]
              ).mean()*100,
              include_groups=False
          )
          .reset_index(name="Accuracy")
    )
```

```python
claude_cat = category_accuracy(claude)
deepseek_cat = category_accuracy(deepseek)
gpt_cat = category_accuracy(gpt)
```
