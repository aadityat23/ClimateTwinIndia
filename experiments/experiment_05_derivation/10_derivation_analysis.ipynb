```python
import pandas as pd
import numpy as np
from pathlib import Path

# ==========================================================
# CHANGE THIS TO YOUR RESULTS DIRECTORY
# ==========================================================

RESULTS = Path(r"E:\ClimateTwinIndia\experiments\experiment_05_derivation\results")

MODELS = [
    "claude",
    "claude_opus",
    "deepseek",
    "gemini",
    "gpt"
]

TOL = 1e-4

rows = []

for model in MODELS:

    file = RESULTS / f"{model}_derivation.csv"

    if not file.exists():
        print(f"Missing: {file}")
        continue

    df = pd.read_csv(file)

    # -----------------------------
    # Metric correctness
    # -----------------------------

    m1 = np.abs(df.GT_Metric1 - df.Pred_Metric1) <= TOL
    m2 = np.abs(df.GT_Metric2 - df.Pred_Metric2) <= TOL

    m3 = (
        df.GT_Metric3.isna()
        |
        (np.abs(df.GT_Metric3 - df.Pred_Metric3) <= TOL)
    )

    metric_correct = m1 & m2 & m3

    decision_correct = df.Decision_Correct

    metric_errors = (~metric_correct).sum()

    absorbed = (
        (~metric_correct)
        &
        (decision_correct)
    ).sum()

    propagated = (
        (~metric_correct)
        &
        (~decision_correct)
    ).sum()

    rows.append({

        "Model": model.replace("_"," ").title(),

        "Questions":
            len(df),

        "Protocol Recovery (%)":
            round(
                100*decision_correct.mean(),
                2
            ),

        "Metric Accuracy (%)":
            round(
                100*metric_correct.mean(),
                2
            ),

        "Metric Errors":
            int(metric_errors),

        "Absorbed":
            int(absorbed),

        "Propagated":
            int(propagated),

        "Absorption Rate (%)":
            round(
                100*absorbed/max(metric_errors,1),
                2
            ),

        "Propagation Rate (%)":
            round(
                100*propagated/max(metric_errors,1),
                2
            )

    })

summary = pd.DataFrame(rows)

summary = summary.sort_values(
    "Protocol Recovery (%)",
    ascending=False
)

print("\n")
print("="*110)
print(summary)
print("="*110)

summary.to_csv(
    RESULTS/"master_isolation_summary.csv",
    index=False
)

print("\nSaved:")
print(RESULTS/"master_isolation_summary.csv")
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

RESULTS = Path("results")

MODELS = {
    "Claude": "claude",
    "Claude Opus": "claude_opus",
    "DeepSeek": "deepseek",
    "Gemini": "gemini",
    "GPT": "gpt"
}

out = []

out.append("="*90)
out.append("CLIMATETWINBENCH - MASTER PAPER STATISTICS")
out.append("="*90)
out.append("")

###############################################################
# Helper
###############################################################

def metric_correct(df):

    m1 = np.isclose(df["GT_Metric1"], df["Pred_Metric1"], atol=1e-4)

    m2 = np.isclose(df["GT_Metric2"], df["Pred_Metric2"], atol=1e-4)

    if "GT_Metric3" in df.columns:

        m3 = (
            df["GT_Metric3"].isna()
            |
            np.isclose(
                df["GT_Metric3"],
                df["Pred_Metric3"],
                atol=1e-4
            )
        )

    else:
        m3 = True

    return m1 & m2 & m3


###############################################################
# Overall model summaries
###############################################################

master = []

for nice_name, model in MODELS.items():

    file = RESULTS / f"{model}_derivation.csv"

    if not file.exists():
        continue

    df = pd.read_csv(file)

    metric = metric_correct(df)

    decision = df["Decision_Correct"]

    total = len(df)

    metric_errors = (~metric).sum()

    absorbed = ((~metric) & decision).sum()

    propagated = ((~metric) & (~decision)).sum()

    protocol = decision.mean()*100

    metric_acc = metric.mean()*100

    absorb_rate = (
        absorbed/metric_errors*100
        if metric_errors else 0
    )

    prop_rate = (
        propagated/metric_errors*100
        if metric_errors else 0
    )

    master.append({
        "Model": nice_name,
        "Questions": total,
        "Protocol Recovery": protocol,
        "Metric Accuracy": metric_acc,
        "Metric Errors": metric_errors,
        "Absorbed": absorbed,
        "Propagated": propagated,
        "Absorption Rate": absorb_rate,
        "Propagation Rate": prop_rate
    })

master_df = pd.DataFrame(master)

out.append("OVERALL MODEL PERFORMANCE")
out.append("-"*90)
out.append(master_df.to_string(index=False))
out.append("")
out.append("")

###############################################################
# Per-category summaries
###############################################################

out.append("="*90)
out.append("PER-CATEGORY SUMMARY")
out.append("="*90)
out.append("")

for nice_name, model in MODELS.items():

    file = RESULTS / f"{model}_derivation.csv"

    if not file.exists():
        continue

    df = pd.read_csv(file)

    metric = metric_correct(df)

    df["MetricCorrect"] = metric

    out.append(f"\n{nice_name}")
    out.append("-"*50)

    summary = []

    for cat, g in df.groupby("subcategory"):

        mc = g["MetricCorrect"]

        dc = g["Decision_Correct"]

        metric_errors = (~mc).sum()

        absorbed = ((~mc) & dc).sum()

        propagated = ((~mc) & (~dc)).sum()

        summary.append({

            "Category":cat,

            "Questions":len(g),

            "MetricAcc":round(mc.mean()*100,2),

            "DecisionAcc":round(dc.mean()*100,2),

            "MetricErrors":metric_errors,

            "Absorbed":absorbed,

            "Propagated":propagated

        })

    out.append(
        pd.DataFrame(summary).to_string(index=False)
    )
    out.append("")

###############################################################
# Propagated failures
###############################################################

out.append("="*90)
out.append("PROPAGATED FAILURES")
out.append("="*90)
out.append("")

for nice_name, model in MODELS.items():

    file = RESULTS / f"{model}_derivation.csv"

    if not file.exists():
        continue

    df = pd.read_csv(file)

    metric = metric_correct(df)

    fail = df[(~metric) & (~df["Decision_Correct"])]

    out.append(f"\n{nice_name}")

    if len(fail)==0:

        out.append("None\n")
        continue

    cols=[
        "id",
        "subcategory",
        "GT_Decision",
        "RecoveredDecision",
        "GT_Metric1",
        "Pred_Metric1",
        "GT_Metric2",
        "Pred_Metric2",
        "GT_Metric3",
        "Pred_Metric3"
    ]

    out.append(
        fail[cols].to_string(index=False)
    )
    out.append("")

###############################################################
# Benchmark totals
###############################################################

out.append("="*90)
out.append("BENCHMARK TOTALS")
out.append("="*90)
out.append("")

example = pd.read_csv(
    RESULTS/"claude_derivation.csv"
)

out.append(f"Questions                  : {len(example)}")
out.append(f"Categories                 : {example.subcategory.nunique()}")

out.append(
    f"Persistent                 : {(example.subcategory=='persistent_regional_anomaly_verification').sum()}"
)

out.append(
    f"Localized                  : {(example.subcategory=='localized_intensification_verification').sum()}"
)

out.append(
    f"Wet                        : {(example.subcategory=='wet_anomaly_consistency_verification').sum()}"
)

out.append(
    f"Compound                   : {(example.subcategory=='compound_state_transition_verification').sum()}"
)

out.append(
    f"Spatial                    : {(example.subcategory=='spatial_coherence_verification').sum()}"
)

out.append(
    f"Reliability                : {(example.subcategory=='reliability_aware_claim_verification').sum()}"
)

out.append("")

###############################################################
# Save
###############################################################

outfile = RESULTS/"paper_numbers.txt"

with open(outfile,"w",encoding="utf8") as f:

    f.write("\n".join(out))

print("="*80)
print("Saved")
print(outfile)
print("="*80)
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

RESULTS = Path("results")
TABLES = RESULTS / "latex_tables"
TABLES.mkdir(exist_ok=True)

MODELS = {
    "Claude": "claude",
    "Claude Opus": "claude_opus",
    "DeepSeek": "deepseek",
    "Gemini": "gemini",
    "GPT-OSS-120B": "gpt"
}


# ---------------------------------------------------------
# Helper
# ---------------------------------------------------------

def metric_correct(df):

    m1 = np.isclose(df.GT_Metric1, df.Pred_Metric1, atol=1e-4)

    m2 = np.isclose(df.GT_Metric2, df.Pred_Metric2, atol=1e-4)

    m3 = (
        df.GT_Metric3.isna()
        |
        np.isclose(
            df.GT_Metric3,
            df.Pred_Metric3,
            atol=1e-4
        )
    )

    return m1 & m2 & m3


# =========================================================
# TABLE 1
# Overall Isolation Results
# =========================================================

rows = []

for pretty, short in MODELS.items():

    file = RESULTS / f"{short}_derivation.csv"

    if not file.exists():
        continue

    df = pd.read_csv(file)

    metric = metric_correct(df)

    decision = df.Decision_Correct

    metric_errors = (~metric).sum()

    absorbed = ((~metric) & decision).sum()

    propagated = ((~metric) & (~decision)).sum()

    rows.append({

        "Model":pretty,

        "Questions":len(df),

        "Protocol Recovery (%)":
            round(decision.mean()*100,2),

        "Metric Accuracy (%)":
            round(metric.mean()*100,2),

        "Metric Errors":metric_errors,

        "Absorbed":absorbed,

        "Propagated":propagated

    })

table1 = pd.DataFrame(rows)

table1.to_latex(

    TABLES/"table1_isolation.tex",

    index=False,

    float_format="%.2f"

)


# =========================================================
# TABLE 2
# Category Breakdown
# =========================================================

rows=[]

for pretty, short in MODELS.items():

    file = RESULTS / f"{short}_derivation.csv"

    if not file.exists():
        continue

    df=pd.read_csv(file)

    metric=metric_correct(df)

    df["MetricCorrect"]=metric

    for cat,g in df.groupby("subcategory"):

        mc=g.MetricCorrect

        dc=g.Decision_Correct

        rows.append({

            "Model":pretty,

            "Category":cat,

            "Metric Accuracy (%)":
                round(mc.mean()*100,2),

            "Decision Accuracy (%)":
                round(dc.mean()*100,2),

            "N":len(g)

        })

table2=pd.DataFrame(rows)

table2.to_latex(

    TABLES/"table2_category_breakdown.tex",

    index=False,

    float_format="%.2f"

)


# =========================================================
# TABLE 3
# Error Absorption
# =========================================================

rows=[]

for pretty, short in MODELS.items():

    file=RESULTS/f"{short}_derivation.csv"

    if not file.exists():
        continue

    df=pd.read_csv(file)

    metric=metric_correct(df)

    decision=df.Decision_Correct

    metric_errors=(~metric).sum()

    absorbed=((~metric)&decision).sum()

    propagated=((~metric)&(~decision)).sum()

    rows.append({

        "Model":pretty,

        "Metric Errors":metric_errors,

        "Absorbed":absorbed,

        "Propagated":propagated,

        "Absorption Rate (%)":
            round(absorbed/max(metric_errors,1)*100,2),

        "Propagation Rate (%)":
            round(propagated/max(metric_errors,1)*100,2)

    })

table3=pd.DataFrame(rows)

table3.to_latex(

    TABLES/"table3_cascade.tex",

    index=False,

    float_format="%.2f"

)


# =========================================================
# TABLE 4
# Propagated Failures
# =========================================================

rows=[]

for pretty, short in MODELS.items():

    file=RESULTS/f"{short}_derivation.csv"

    if not file.exists():
        continue

    df=pd.read_csv(file)

    metric=metric_correct(df)

    fail=df[(~metric)&(~df.Decision_Correct)]

    for _,r in fail.iterrows():

        rows.append({

            "Model":pretty,

            "Question":r.id,

            "Category":r.subcategory,

            "GT":r.GT_Decision,

            "Predicted":r.RecoveredDecision

        })

table4=pd.DataFrame(rows)

table4.to_latex(

    TABLES/"table4_failures.tex",

    index=False

)


# =========================================================
# TABLE 5
# Benchmark Composition
# =========================================================

example=pd.read_csv(
    RESULTS/"claude_derivation.csv"
)

rows=[]

for cat,g in example.groupby("subcategory"):

    rows.append({

        "Category":cat,

        "Questions":len(g)

    })

table5=pd.DataFrame(rows)

table5.to_latex(

    TABLES/"table5_benchmark.tex",

    index=False

)


print("="*70)
print("Generated:")
print("="*70)

for f in sorted(TABLES.glob("*.tex")):
    print(f.name)

print("="*70)
```
